```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import utils
```

```python
df = sns.load_dataset('fmri')
df['signal_err'] = np.abs(np.ones(len(df['signal'])))*0.1
sns.lineplot(df, x='timepoint', y='signal', errorbar='se')
utils.plot.lineplot(df, x='timepoint', y='signal', yerr='signal_err', weighted=False, label='not weighted')
l = utils.plot.lineplot(df, x='timepoint', y='signal', yerr='signal_err', weighted=True, label='weighted')
plt.legend()
plt.show()
utils.plot.extract_data(l)
```

```python
utils.plot.vec([1,2], origin=[2,1])
utils.plot.set_aspect('equal_square')
plt.grid(which='both')
plt.show()
```

```python
mean = np.array([1,2])
U = 1/(2**0.5)*np.array([[1,-1],[1,1]])
S = np.diag(np.array([3,1]))**2
cov = U @ S @ U.T
x = np.random.multivariate_normal(mean,cov,size=100)

fig, axes = utils.plot.subplots(1, 2, sharex=True, sharey=True)

utils.plot.regplot(x[:,0], x[:,1], text_kwargs={'loc': 'upper left'}, ax=axes[0])

sns.regplot(x=x[:,0], y=x[:,1], ax=axes[1])

plt.show()
```

```python
fig, axes = utils.plot.subplots(1,3,keep_shape=True)
x, y = np.linspace(0,5), np.linspace(1,6)
for i in range(axes.shape[0]):
    for j in range(axes.shape[1]):
        axes[i,j].plot(x, y+j)
plt.show()

fig, axes = utils.plot.subplots(1,3,keep_shape=False)
x, y = np.linspace(0,5), np.linspace(1,6)
for j in range(axes.shape[0]):
    axes[j].plot(x, y+j)
plt.show()
```
